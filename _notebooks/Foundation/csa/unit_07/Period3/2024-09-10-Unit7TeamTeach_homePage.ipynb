        <!-- Bubbles will be added here dynamically -->
    </div>
    <div>
        <button onclick="addBubble()">Add Bubble</button>
        <button onclick="removeBubble()">Remove Bubble</button>
        <button onclick="shuffleBubbles()">Shuffle Bubbles</button>
    </div>
    <script>
        let bubbles = [];
        function addBubble() {
            let bubble = document.createElement('div');
            bubble.classList.add('bubble');
            bubble.style.backgroundColor = getRandomColor();
            document.getElementById('bubbleContainer').appendChild(bubble);
            bubbles.push(bubble);
        }
        function removeBubble() {
            if (bubbles.length > 0) {
                let bubble = bubbles.pop();
                bubble.remove();
            }
        }
        function shuffleBubbles() {
            let container = document.getElementById('bubbleContainer');
            for (let i = container.children.length; i >= 0; i--) {
                container.appendChild(container.children[Math.random() * i | 0]);
            }
        }
        function getRandomColor() {
            let letters = '0123456789ABCDEF';
            let color = '#';
            for (let i = 0; i < 6; i++) {
                color += letters[Math.floor(Math.random() * 16)];
            }
            return color;
        }
    </script>
